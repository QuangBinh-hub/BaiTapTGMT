```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ng dn
my_path='D:/TKWeb/6_TranQuangBinh_21DTH2_Buoi7/Image/1.jpg'
im = Image.open(my_path)
plt.imshow(im)
```

```python
#chuyn nh xm
im_gray=im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title("Biu  Histogram ca nh")
plt.xlabel("Gi tr ca pixel")
plt.ylabel("S lng")
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray,origin="image")
plt.show()
```
