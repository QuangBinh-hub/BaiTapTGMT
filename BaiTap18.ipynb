```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Đọc ảnh
image = Image.open('D:/TKWeb/6_TranQuangBinh_21DTH2_Buoi7/Image/1.png')

# Chuyển sang ảnh xám
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)

#Thực hiện openning để loại bỏ nhiễu và nối đối tượng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((5,5)))

# Đánh nhãn
labeled_image, num_features = ndimage.label(opened_image)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Đọc ảnh
image = Image.open('D:/TKWeb/6_TranQuangBinh_21DTH2_Buoi7/Image/1.png')

# Chuyển sang ảnh xám
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)

#Thực hiện openning để loại bỏ nhiễu và nối đối tượng
closed_image = ndimage.binary_closing(binary_image, structure=np.ones((7,7)))

# Đánh nhãn
labeled_image, num_features = ndimage.label(closed_image)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,3)
plt.imshow(closed_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```
